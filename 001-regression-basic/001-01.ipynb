```python
import numpy as np

x = np.array([
    0.62, 0.64, 0.65, 0.71, 0.71, 0.72, 0.72, 0.73, 0.73, 0.75, 0.76, 0.76, 0.77, 0.78, 0.78, 0.78, 0.80, 0.80, 0.82, 0.82, 0.83, 0.83, 0.84, 0.85, 0.85, 0.85, 0.85, 0.88, 0.88, 0.90, 0.94, 0.94, 0.96, 1.00, 1.00
])

y = np.array([
    0.242, 0.29, 0.34, 0.363, 0.43, 0.45, 0.5, 0.39, 0.45, 0.5, 0.475, 0.5, 0.5, 0.34, 0.6, 0.6, 0.7, 0.7, 0.61, 0.65, 0.575, 0.685, 0.62, 0.68, 0.7, 0.725, 0.72, 0.714, 0.85, 1, 0.92, 0.955, 0.925, 0.975, 0.95
])

import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x.reshape(-1, 1), y)

print(lr.coef_, lr.intercept_)
```

```python
coef = [1.0, 2.0, 3.3, 0.0]
intercept = [-0.25, -1.0, -2.0, 0.6]

# x   (  )
x_line = np.linspace(0.6, 1.0, 100)  # 0.6 1.0 100 

y_line = [
    coef[0] * x_line + intercept[0],
    coef[1] * x_line + intercept[1],
    coef[2] * x_line + intercept[2],
    coef[3] * x_line + intercept[3]
]

#  
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
plt.plot(x_line, y_line[0], color='red', linewidth=2, label=f"y = {coef[0]}x + {intercept[0]}")
plt.plot(x_line, y_line[1], color='green', linewidth=2, label=f"y = {coef[1]}x + {intercept[1]}")
plt.plot(x_line, y_line[2], color='blue', linewidth=2, label=f"y = {coef[2]}x + {intercept[2]}")
plt.plot(x_line, y_line[3], color='purple', linewidth=2, label=f"y = {coef[3]}x + {intercept[3]}", linestyle=':')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
