```python
import numpy as np

x1 = np.array([
    0.62, 0.64, 0.65, 0.71, 0.71, 0.72, 0.72, 0.73, 0.73, 0.75, 0.76, 0.76, 0.77, 0.78, 0.78, 0.78, 0.80, 0.80, 0.82, 0.82, 0.83, 0.83, 0.84, 0.85, 0.85, 0.85, 0.85, 0.88, 0.88, 0.90, 0.94, 0.94, 0.96, 1.00, 1.00
])

x2 = np.array([
    0.61, 0.66, 0.65, 0.67, 0.66, 0.72, 0.75, 0.67, 0.74, 0.75, 0.75, 0.76, 0.73, 0.73, 0.79, 0.81, 0.78, 0.79, 0.82, 0.76, 0.80, 0.84, 0.82, 0.82, 0.86, 0.86, 0.86, 0.87, 0.89, 1.00, 0.95, 0.95, 0.99, 0.98, 0.93
])

y = np.array([
    0.242, 0.29, 0.34, 0.363, 0.43, 0.45, 0.5, 0.39, 0.45, 0.5, 0.475, 0.5, 0.5, 0.34, 0.6, 0.6, 0.7, 0.7, 0.61, 0.65, 0.575, 0.685, 0.62, 0.68, 0.7, 0.725, 0.72, 0.714, 0.85, 1, 0.92, 0.955, 0.925, 0.975, 0.95
])

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 3D  
scatter = ax.scatter(x1, x2, y, c=y, cmap='viridis', s=50, alpha=0.8)

#   
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('3D Scatter Plot: x1, x2 vs y')

#  
cbar = plt.colorbar(scatter)
cbar.set_label('y ')

#  
ax.grid(True, alpha=0.3)

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

#   (2D -> 2D  )
X = np.column_stack((x1, x2))

#      
model = LinearRegression()
model.fit(X, y)

#   
print(f" (coefficients): a1={model.coef_[0]:.4f}, a2={model.coef_[1]:.4f}")
print(f" (intercept): {model.intercept_:.4f}")
#  
print(f"\n: y = {model.intercept_:.4f} + {model.coef_[0]:.4f}x1 + {model.coef_[1]:.4f}x2")

#  
y_pred = model.predict(X)
print(f"R : {r2_score(y, y_pred):.4f}")
print(f"MSE: {mean_squared_error(y, y_pred):.6f}")


# 3D 
fig = plt.figure(figsize=(12, 5))

# :  
ax1 = fig.add_subplot(121, projection='3d')
scatter1 = ax1.scatter(x1, x2, y, c='blue', s=50, alpha=0.8, label=' ')
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('y')
ax1.set_title('Real Data')
ax1.grid(True, alpha=0.3)

# :   
ax2 = fig.add_subplot(122, projection='3d')

#   
x1_range = np.linspace(x1.min(), x1.max(), 20)
x2_range = np.linspace(x2.min(), x2.max(), 20)
X1, X2 = np.meshgrid(x1_range, x2_range)
Y_pred = model.intercept_ + model.coef_[0] * X1 + model.coef_[1] * X2

ax2.plot_surface(X1, X2, Y_pred, alpha=0.3, color='red', label='Regression Plane')
scatter2 = ax2.scatter(x1, x2, y, c='blue', s=50, alpha=0.8, label='Real Data')
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_zlabel('y')
ax2.set_title('Regression Plane and Data')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```
